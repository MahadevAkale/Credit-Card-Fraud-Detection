```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
# Impute missing values in X_train and X_test with the median
imputer_X = SimpleImputer(strategy='median')
X_train = imputer_X.fit_transform(X_train)
X_test = imputer_X.transform(X_test)

# Impute missing values in y_train and y_test with the most frequent value
imputer_y = SimpleImputer(strategy='most_frequent')
y_train = imputer_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test = imputer_y.transform(y_test.values.reshape(-1, 1)).ravel()
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"\n Accuaracy: {accuracy_score(y_test, y_pred)}")
    print(f"\n Precision: {precision_score(y_test, y_pred)}")
    print(f"\n Recall: {recall_score(y_test, y_pred)}")
    print(f"\n F1 Score: {f1_score(y_test, y_pred)}")
```

```python
import tkinter as tk
from tkinter import messagebox
import joblib
import numpy as np

# Load the model
model = joblib.load("credit_card_model.pkl")

# Create the main window
root = tk.Tk()
root.title("Credit Card Transaction Classifier")
root.geometry("500x300")

# Create a label for the app
label = tk.Label(root, text="Enter 29 Transaction Data Values (comma-separated):", font=("Arial", 14))
label.pack(pady=10)

# Create an Entry widget for the user to input 29 comma-separated values
entry = tk.Entry(root, width=60)
entry.pack(pady=20)

# Function to classify transaction
def classify_transaction():
    try:
        # Get input data from the user as a single string
        input_data = entry.get()
        
        # Split the string by commas and convert the values to floats
        transaction_data = [float(value) for value in input_data.split(',')]
        
        # Check if 29 values are provided
        if len(transaction_data) != 29:
            raise ValueError("Fraudulent Transaction")
        
        # Make prediction using the loaded model
        pred = model.predict([transaction_data])
        
        # Display result
        if pred[0] == 0:
            result = "Normal Transaction"
        else:
            result = "Fraudulent Transaction"
        
        # Show result in a message box
        messagebox.showinfo("Prediction", result)
    
    except ValueError as e:
        messagebox.showerror("Fraudulent Transaction", str(e))

# Create a button to classify the transaction
classify_button = tk.Button(root, text="Predict", command=classify_transaction)
classify_button.pack(pady=20)

# Run the main loop
root.mainloop()
```
